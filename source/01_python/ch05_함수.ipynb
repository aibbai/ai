````python
# args 앞의 * : 매개변수들(인수들)이 args 튜플로 전달
def add(*args) :
    "전달받은 매개변수들의 합을 return. 매개변수가 없을 경우 0을 return"
    # print(args) # ()
    tot = 0 # 누적변수
    for num in args:
        tot += num # tot = tot + num
    return tot
````

## 3.7 인수의 언패킹
### 튜플 인수 언패킹
- 패킹 : 함수 정의시 매개변수 앞에 *붙임 (튜플 매개변수)
- 언패킹 : 함수 호출 시 매개변수 앞에 *붙임 (리스트나 튜플이 각자 요소별 매개변수로)

````python
def add(*args) :
    tot = 0
    for num in args:
        tot += num
    return tot
````

````python
# ex5. safe_index(ch03.py)를 람다식으로
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst:
        return lst.index(item)
    return -1
````

````python
def map_template(func, L=[]):
    result = []
    for item in L:
        result.append(func(item))
    return result
````
