## 1.4 인덱싱과 인덱스 찾기
- 변수[idx] : 인덱스를 이용한 접근(없는 인덱스 사용시 에러)
- count(찾을 데이터) : 해당 찾을 데이터가 나오는 갯수 반환
- index(찾을 데이터) : 0번째부터 해당 찾을 데이터가 몇번째 index에 있는지 반환
     (index는 count 다음에 찾기 (없으면 에러니까) or in 연산자 써도 됨 or safe_index)
- index(찾을 데이터, start) : start번째부터 해당 찾을 데이터가 몇번째 index에 있는지 반환

```python
numbers = [1, 2, 3, [1, 2, 3], 1, 2, 3]
```

```python
if numbers.count(9):
    print(numbers.index(9))
else: 
    print('찾을 수 없음')
```

```python
if 9 in numbers:
    print(numbers.index(9))
else:
    print('찾을 수 없음')
```

```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 반환하는 함수
    item 요소가 없으면 -1를 반환
    '''
    if item in lst:
        return lst.index(item)
    return -1
```

## 3.2 딕셔너리 item 추가, 수정, 제거

```python
my_favorite = {'fruit':'apple', 'number':12, 'sports':'yoga'}
# 한 item씩 추가
my_favorite['people'] = '홍길동'

print('1. 한 item 추가 후 :', my_favorite)
# 여러 item을 한꺼번에 추가
my_favorite.update({'subject':'수학', 'major':'AI'})
print('2. 여러 item 추가 후 :', my_favorite)

# 한 item씩 수정
my_favorite['fruit'] = 'banana'
print('3. 한 item 수정 후 :', my_favorite)
# 여러 item을 한꺼번에 수정
# my_favorite.update({'fruit':'kiwi', 'number':-99})
my_favorite.update(fruit='kiwi', number=-99)
print('4. 여러 item 수정 후', my_favorite)

# item 제거 방법1 : del 변수 ['삭제할 key'] - 없는 key일 경우 에러
del my_favorite['fruit']
if 'fruit' in my_favorite:
    del my_favorite['fruit']
else:
    print('5. fruit 삭제하여 존재하지 않음')
# item 제거 방법2 : 변수.pop('삭제할 key') - 해당 key값을 반환하고 제거 (없는 key일 경우 에러)
print('6. 제거한 데이터 :', my_favorite.pop('number'))
if 'number' in my_favorite:
    print('7. 제거한 데이터 :', my_favorite.pop('number'))
else:
    print('7. number 삭제하여 존재하지 않음')
print(my_favorite)
```

## 3.3 출력하기
- for문 이용

```python
my_tuple = ('홍길동', 30, '0109999999') # {0:'홍길동', 1:30, 2:'010999999'}
for data in my_tuple: 
    print(data, end='\t')
print() # 개행
for idx in range(len(my_tuple)): #[0,1,2]
    print("{}번째 : {}\t".format(idx, my_tuple[idx]), end="")
print()
for idx, value in enumerate(my_tuple): # (0,홍길동), (1,30),(2,'01099999999')
    print(f"{idx}번째 : {value}\t", end="")
```

# 5절. enumerate
- 반복자 또는 순서객체로 반복문을 처리할 때 사용하는 함수
- 리스트, 튜플, 셋을 for문에 사용시

```python
list(enumerate(names))
```

```python
for no, name in enumerate(names):
    print("{} : {}".format(no, name))
```

```python
names = ['길동', '길자', '순자']
tels = ['010-9999-9999',
       '010-8888-8888',
       '010-7777-7777']
adds = ['서울', '인천', '판교']
for idx in range(len(names)): # [0,1,2]
    print("{}. {}의 연락처는 {}이고, {} 살아요".format(idx+1,
                                            names[idx],
                                            tels[idx],
                                            adds[idx]))
```

```python
list(zip(names, tels, adds))
```

```python
for name, tel, add in zip(names, tels, adds): 
    print(f"{name}의 연락처는 {tel}이고, {add}에 살아요")
```

```python
list(enumerate(zip(names, tel, adds)))
```

```python
for idx, (name, tel, add) in enumerate(zip(names, tels, adds)):
    print(f"{idx+1}. {name}의 연락처는 {tel}이고, {add}에 살아요")
```

```python
# names, tels, adds를 딕셔너리 리스트로 (리스트 안에 딕셔너리)
members = []
for name, tel, add in zip(names, tels, adds):
    member = {'name': name, 'tel':tel, 'address':add}
    members.append(member)
members
```

```python
members = []
for idx, (name, tel, add) in enumerate(zip(names, tels, adds)):
    member = {'순위':idx+1, '이름':name, '전화':tel, '주소':add}
    members.append(member)
members
```
