정규표현식(regex)
```
    1. 참조 : https://ko.wikipedia.org/wiki/정규_표현식
    2. 간략문법
        \d : (숫자와 매치. [0-9]와 동일)
        \D : (숫자가 아닌 문자)
        \w : (영문자나 숫자. [a-zA-Z0-9]와 동일)
        /W : (영문자나 숫자가 아닌 문자)
        . : (문자)
        \. : . #문자
        \s : () #소괄호
        {n,} : n회 반복
        {2,4} : 2~4회 반복 #4,2는 안됨 (작은 숫자가 앞으로)
        ? : (0번이나 1번 의미) 
        + : (1번 이상 반복)
        * : (0번 이상 반복)
    3. 정규표현식 연습장 : https://regexr.com/
        ex1. 전화번호 패턴 : 0d{1,2}[\-)\s\.]\d{3,4}[\-\s\.]\d{4}
        ex2. 이메일 패턴 : \w[a-z0-9_]+@\w+(\.\w+){1,2}
    
```

````python
data = "이름:홍길동, 주소:신림동, 전화:02-888-8888, 02-741-8745, 02)888-7878 \
핸드폰:010-9999-9999, 이메일:abc@lg.co.kr, abc@naver.com \
특징:서해번쩍 동해번쩍, 혈액형:B ㅋㅋ"
print(data)
````

````python
import re
tel_pattern = r"0\d{1,2}[\-)\s\.]\d{3,4}[\-\s\.]\d{4}"
tels = re.findall(tel_pattern, data) # 해당 패턴을 다 찾아줘. 없으면 빈 list([])
tels
````

````python
# 정규표현식 패턴과 처음 match되는 것을 찾아줘
firstTel = re.search(tel_pattern, data)
firstTel
````

````python
print(firstTel.group(0)) #group(0)하면 전부 다 가져옴
print(data[19:30])
````

````python
phone_pattern = r"(\d{2,3})-(\d{3,4})-(\d{4})"
phone = re.search(phone_pattern, data)
print(phone.group(0))
print(phone.group(1))
print(phone.group(2))
print(phone.group(3))
````

````python
email_pattern = r"\w[a-z0-9_]+@\w+(?:\.[a-z]+){1,2}"
print(re.findall(email_pattern, data))
print(re.search(email_pattern, data).group(0))
````

````python
# 특정 문자외 없애기
re.sub(r"[^가-힣0-9\-a-z@]", " ", data)
````

## 4.2 format() 문법

- "{[인덱스]:[공백대체문자][정렬방식][자릿수][타입]}".format(변수명)
    * 인덱스 : format 함수 안의 매개변수 순서
    * 공백대체문자는 정렬방식과 자릿수를 명시하여야 함, 자릿수를 여유롭게
    * 정렬방식 : <(왼쪽정렬), >(오른쪽정렬), ^(가운데정렬) #안하면 숫자는 오른쪽, 문자는 오른쪽이 default
    * 정수 자릿수 : 출력할 자릿수
    * 실수 자릿수 : 전체출력 될 자릿수, 변수 중 출력할 자릿수
    * 타입 : d(10진수), f(실수 ex:4.1f:1은 소수점자리수), s(문자), o(8진수), x(16진수), b(2진수)

````python
name = input("이름은?")
age = int(input("나이는?"))
````

````python
print("[{0: ^10d}]살 [{1:_>10s}]님 안녕하세요".format(age, name))
print(f"[{age:*<10f}]살 [{name:_>10.2s}]님 안녕하세요")
````

````python
a = 2.755
sign = '+'
width = 10
digits = 3
print("[{:+10.3}]".format(a)) # 부호 넣고 싶으면(+) 자릿수 앞에, 그리고 .3은 열 자리 중에 세 자리만 
print("[{:{sign}{width}.{digits}}]".format(a, 
                                           sign=sign,
                                           width=width,
                                           digits=digits))
````
